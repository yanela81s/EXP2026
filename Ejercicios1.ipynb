Los teléfonos de una empresa tienen el siguiente formato prefijo-número-extensión donde el prefijo es el código del país +34, y la extensión tiene dos dígitos (por ejemplo +34-913724710-56). Escribir un programa que pregunte por un número de teléfono con este formato y muestre por pantalla el número de teléfono sin el prefijo y la extensión.

```python
telefono = input("Escribe tu número de teléfono: ")
```

```python
telefono = input("Escribe tu número de teléfono: ")

primer_guion = telefono.find("-")
segundo_guion = telefono.find("-", primer_guion + 1)

numero = telefono[primer_guion + 1 : segundo_guion]

print("Número sin prefijo ni extensión:", numero)

```

Escribir un programa que pregunte por consola el precio de un producto en dólares con dos decimales y muestre el valor en soles. Posteriormente, pedir al usuario que indique la posición del punto decimal, para que el resultado sea el siguiente:
Introduce el precio del producto con dos decimales: 12.45

En soles 45.6915

Indicar posición del punto: 2

45 soles con 6915 céntimos

```python
precio_dolares = input("Introduce el precio del producto con dos decimales ")
precio_dolares = float(precio_dolares)

precio_soles = precio_dolares * 3.36
print("En soles", precio_soles)
```

```python
from os import posix_fadvise
posicion_decimal = int(input("Indicar posición del punto "))
precio_texto = str(precio_soles)

soles = precio_texto[:posicion_decimal]
centimos = precio_texto[posicion_decimal+1:]

print(soles, "soles con", centimos, "céntimos")
```

```python
precio_dolares = float(input("Introduce el precio del producto con dos decimales "))
precio_soles = precio_dolares * 3.36

precio_texto = str(precio_soles)

posicion_punto = precio_texto.find(".")

soles = precio_texto[:posicion_punto]
centimos = precio_texto[posicion_punto+1:]

print(soles, "soles con", centimos, "céntimos")
```
